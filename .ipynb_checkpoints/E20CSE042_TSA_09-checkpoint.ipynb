```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from sklearn.metrics import mean_squared_error
```

```python
df = pd.read_csv('perrin-freres-monthly-champagne.csv')
df.head()
```

```python
df.rename(columns={"Perrin Freres monthly champagne sales millions ?64-?72": "Sales"}, inplace=True)
df.head()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df.index, df.Sales)
plt.title('Monthly Champagne Sales')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.show()
```

```python
df.isnull().sum().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum().sum()
```

```python
df.isnull().sum()
```

```python
df.sort_index(inplace=True)
```

```python
df.isnull().sum()
```

```python
result = seasonal_decompose(x=df["Sales"], model='multiplicative',period=1)
result.plot()
plt.show()
```

```python
result = adfuller(df.Sales)
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))
```

```python
diff = df.Sales.diff().dropna()

result = adfuller(diff)
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10,5))
plot_acf(diff, ax=ax[0])
plot_pacf(diff, ax=ax[1])
plt.show()
```

```python
order = (1, 1, 0)
seasonal_order = (1, 0, 0, 12)

model = SARIMAX(df.Sales, order=order, seasonal_order=seasonal_order)
results = model.fit()

print(results.summary())
```

```python
residuals = results.resid

fig, ax = plt.subplots(1, 2, figsize=(10,5))
residuals.plot(ax=ax[0])
ax[0].set_title('Residuals')
ax[0].set_xlabel('Date')
ax[0].set_ylabel('Sales')
plot_acf(residuals, ax=ax[1])
plt.show()

sns.kdeplot(residuals)
plt.title('Density Plot of Residuals')
plt.xlabel('Residuals')
plt.show()
```

```python
forecast = results.forecast(steps=48)

fig, ax = plt.subplots(figsize=(10,5))
df.Sales.plot(ax=ax)
forecast.plot(ax=ax)
ax.legend(['Actual', 'Forecast'])
ax.set_xlabel('Date')
ax.set_ylabel('Sales')
plt.show()
```

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing

model = ExponentialSmoothing(df.Sales, seasonal_periods=12, trend='add', seasonal='add')
results = model.fit()

print(results.summary())
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np

# SARIMA
sarima_mse = mean_squared_error(df.Sales[-48:], forecast)
sarima_rmse = np.sqrt(sarima_mse)
sarima_mae = mean_absolute_error(df.Sales[-48:], forecast)

# ETS
ets_mse = mean_squared_error(df.Sales[-48:], forecast)
ets_rmse = np.sqrt(ets_mse)
ets_mae = mean_absolute_error(df.Sales[-48:], forecast)

print('SARIMA MSE:', sarima_mse)
print('SARIMA RMSE:', sarima_rmse)
print('SARIMA MAE:', sarima_mae)

print('ETS MSE:', ets_mse)
print('ETS RMSE:', ets_rmse)
print('ETS MAE:', ets_mae)
```
