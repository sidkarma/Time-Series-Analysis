```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
df = pd.read_csv("cab_rides.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df['price'].fillna((df['price'].mean()), inplace=True)
```

```python
df = df[['distance', 'cab_type', 'time_stamp', 'destination', 'source', 'price',
       'surge_multiplier', 'name']]
df.head()
```

```python
# Plot the line chart
plt.figure(figsize=(12,6))
plt.plot(df.index, df['price'], color='blue')
plt.title('Cab Rides Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()

# Plot the histogram
plt.figure(figsize=(12,6))
plt.hist(df['price'], bins=50, color='blue')
plt.title('Distribution of Cab Rides Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
df["rounded_timestamp"] = df["time_stamp"] / 1000
df["rounded_timestamp"] = df["rounded_timestamp"].apply(np.floor)

df["date"] = df["rounded_timestamp"].apply(lambda x : datetime.fromtimestamp(x).date())
df["time"] = df["rounded_timestamp"].apply(lambda x: datetime.fromtimestamp(x).time())
df['weekday'] = df['date'].apply(lambda x: x.weekday())
df["weekday"] = df["weekday"].map({0: 'Monday', 1: 'Tuesday', 2: 'Wednesday', 3: 'Thursday', 4: 'Friday', 5: 'Saturday', 6: 'Sunday'})
df['hour'] = df['time'].apply(lambda time: time.hour)

df.loc[(df.hour >= 6) & (df.hour < 12) , 'time_of_day'] = 'Morning'
df.loc[(df.hour >= 12) & (df.hour < 15) , 'time_of_day'] = 'Afternoon'
df.loc[(df.hour >= 15) & (df.hour < 18) , 'time_of_day'] = 'Evening'
df.loc[(df.hour >= 18) | (df.hour < 6) , 'time_of_day'] = 'Night'

df = df[['distance', 'cab_type', 'time_stamp', 'destination', 'source', 'price',
       'surge_multiplier', 'name', 'date', 'time', 'weekday', 'hour', 'time_of_day']]

df.head()
```

```python
df['cab_type'].value_counts()
```

```python
corr = df.corr()

# Visualize the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix for Cab Rides Data')
plt.show()
```

```python
# Square Root Transform

df['price_sqrt'] = np.sqrt(df['price'])

# Plot the transformed df
plt.figure(figsize=(12,6))
plt.plot(df.index, df['price_sqrt'], color='blue')
plt.title('Cab Rides Prices (Square Root Transformed)')
plt.xlabel('Date')
plt.ylabel('Square Root of Price')
plt.show()
```

```python
plt.hist(df['price_sqrt'])
plt.show()
```

```python
# Log Transform

df['price_log'] = np.log(df['price'])

# Plot the transformed data
plt.figure(figsize=(12,6))
plt.plot(df.index, df['price_log'], color='blue')
plt.title('Cab Rides Prices (Log Transformed)')
plt.xlabel('Date')
plt.ylabel('Log of Price')
plt.show()
```

```python
plt.hist(df['price_log'])
plt.show()
```

```python
# Box-Cox Transform
```

```python
from scipy.stats import boxcox

df['price_boxcox'] = boxcox(df['price'], lmbda=0.0)

# Plot the transformed data
plt.figure(figsize=(12,6))
plt.plot(df.index, df['price_boxcox'], color='blue')
plt.title('Cab Rides Prices (Box-Cox Transformed)')
plt.xlabel('Date')
plt.ylabel('Box-Cox of Price')
plt.show()
```

```python
plt.hist(df['price_boxcox'])
plt.show()
```
